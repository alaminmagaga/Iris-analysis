```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a sample dataset
data = {
    'StudentID': [1, 2, 3, 4, 5,5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve','David'],
    'MathScore': [85, 92, 78, 88, 95,24],
    'ScienceScore': [90, 88, 76, 85, 92,13]
}
```

```python
#Create a DataFrame from the dataset
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#check the shape of the data
df.shape
```

```python
#check foforr duplicate values in the  rows
df.duplicated()
```

```python
#check for missing values
df.isnull().sum()
```

```python
# Calculate additional statistics
mean_math_score = df['MathScore'].mean()
mean_science_score = df['ScienceScore'].mean()
print(mean_math_score)
print(mean_science_score)
```

```python
df.describe()
```

```python

#check for the counts of the values
df.Name.value_counts()
```

```python
# Create a new column for the maths and science average score for each student
df['AverageScoreMC'] = (df['MathScore'] + df['ScienceScore']) / 2
```

```python
#delete the student id column
df.drop("StudentID", axis=1, inplace=True)
```

```python
df.head()
```

```python
# Generate a bar chart to visualize the average scores
plt.figure(figsize=(8, 5))
sns.barplot(x='Name', y='AverageScoreMC', data=df)
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.title('Average Scores of Students')
plt.show()
```
